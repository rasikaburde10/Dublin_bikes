```python
import pandas as pd
df_pre['DATE'] = pd.to_datetime(df_pre['DATE'])
df_wea_final['datetime'] = pd.to_datetime(df_wea_final['datetime'])
# Now check the dtypes again
print(df_pre['DATE'].dtype)
print(df_wea_final['datetime'].dtype)
```

```python
# unique values that may not be date-like
print(df_wea_final['datetime'].unique())
# Try to find non-date entries
non_date_values = df_wea_final[~df_wea_final['datetime'].str.match(r'\d{4}-\d{2}-\d{2}')]
print(non_date_values)
```

```python
import pandas as pd
df_wea_final['datetime'] = pd.to_datetime(df_wea_final['datetime'], errors='coerce')
# Check the data types to confirm the conversion
print(df_wea_final.dtypes)
```

```python
# Convert columns to float
numeric_columns = ['tempmax', 'tempmin', 'temp', 'precip', 'windgust', 'humidity']
for column in numeric_columns:
    df_wea_final[column] = pd.to_numeric(df_wea_final[column], errors='coerce')
```

```python
# Check for NaN values after conversion
print(df_wea_final[numeric_columns].isna().sum())
```

```python
# Fill NaN values with the mean of the column
for column in ['tempmax', 'tempmin', 'temp', 'precip', 'humidity']:
    mean_value = df_wea_final[column].mean()
    df_wea_final[column].fillna(mean_value, inplace=True)

df_wea_final.drop('windgust', axis=1, inplace=True)
print(df_wea_final.describe())
```
