```python
#whole new function
import pandas as pd

def calculate_bike_usage(df, interval):
    """
    Calculate bike usage statistics for specified intervals.

    :param df: DataFrame with bike data.
    :param interval: Time interval for grouping ('D' for daily, 'W' for weekly, 'M' for monthly, 'H' for hourly).
    :return: DataFrame with bike usage statistics for each interval.
    """

    df['LAST UPDATED'] = pd.to_datetime(df['LAST UPDATED'])
    df = df.sort_values('LAST UPDATED')
    df['interval'] = df['LAST UPDATED'].dt.to_period(interval)
    # Calculate the difference in 'AVAILABLE BIKES' between each row and the next
    df['diff'] = df.groupby(['STATION ID', 'interval'])['AVAILABLE BIKES'].diff(-1)
    # A positive diff means bikes were returned (bikes_in), a negative diff means bikes were taken (bikes_out)
    df['bikes_in'] = df['diff'].apply(lambda x: -x if x < 0 else 0)
    df['bikes_out'] = df['diff'].apply(lambda x: x if x > 0 else 0)
    # Aggregate the sums of 'bikes_in' and 'bikes_out' for each interval
    bike_usage = df.groupby('interval').agg({'bikes_in': 'sum', 'bikes_out': 'sum'})
    # Calculate the total bike usage as the sum of 'bikes_in' and 'bikes_out'
    bike_usage['total_usage'] = bike_usage['bikes_in'] + bike_usage['bikes_out']

    return bike_usage.reset_index()

```

```python
def calculate_station_usage(df, interval):
    """
    Calculate bike usage statistics for each station over specified intervals.

    :param df: DataFrame with bike data.
    :param interval: Time interval for grouping ('D' for daily, 'W' for weekly, 'M' for monthly, 'H' for hourly).
    :return: DataFrame with bike usage statistics for each station and interval.
    """

    df['LAST UPDATED'] = pd.to_datetime(df['LAST UPDATED'])


    df = df.sort_values('LAST UPDATED')


    df['diff'] = df.groupby(['STATION ID', 'LAST UPDATED'])['AVAILABLE BIKES'].diff().fillna(0)


    df['bikes_in'] = df['diff'].apply(lambda x: x if x > 0 else 0)
    df['bikes_out'] = df['diff'].apply(lambda x: -x if x < 0 else 0)

    df['period'] = df['LAST UPDATED'].dt.to_period(interval)
    station_usage = df.groupby(['STATION ID', 'period']).agg({'bikes_in': 'sum', 'bikes_out': 'sum'})


    station_usage['total_usage'] = station_usage['bikes_in'] + station_usage['bikes_out']

    return station_usage.reset_index()


station_daily_usage = calculate_station_usage(df_pre, 'D')
print(station_daily_usage.head())


```

```python
def calculate_derived_features(df):

    df['LAST UPDATED'] = pd.to_datetime(df['LAST UPDATED'])


    df_sorted = df.sort_values(by=['STATION ID', 'LAST UPDATED'])


    df_sorted['bikes_out'] = -df_sorted.groupby('STATION ID')['AVAILABLE BIKES'].diff().fillna(0)

    df_sorted['bikes_in'] = df_sorted.groupby('STATION ID')['AVAILABLE BIKES'].diff(periods=-1).fillna(0)

    df_sorted['turnover_rate'] = df_sorted['bikes_out'] / df_sorted['BIKE STANDS']


    df_sorted['is_idle'] = (df_sorted['bikes_in'] == 0) & (df_sorted['bikes_out'] == 0)


    df_sorted['hour'] = df_sorted['LAST UPDATED'].dt.hour
    peak_usage = df_sorted.groupby(['STATION ID', 'hour'])['bikes_out'].sum().reset_index()


    peak_hours = peak_usage.loc[peak_usage.groupby('STATION ID')['bikes_out'].idxmax()]

    return df_sorted, peak_hours


```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt


df_daily_pre = pd.read_csv('daily_usage_pre_everyday.csv')
df_daily_cov = pd.read_csv('daily_usage_cov_everyday.csv')

# Convert the 'interval' columns to datetime

df_daily_pre['interval'] = pd.to_datetime(df_daily_pre['interval'])
df_daily_cov['interval'] = pd.to_datetime(df_daily_cov['interval'])

# Create lagged features for the pre-COVID data
lags = 7  # Number of days for lagged features
for i in range(1, lags + 1):
    df_daily_pre[f'lag_{i}'] = df_daily_pre['total_usage'].shift(i)

# Remove rows with NaN values caused by shifting to create lags in the pre-COVID data
df_daily_pre = df_daily_pre.dropna()

# Prepare the features and target variable for the pre-COVID data
X_pre = df_daily_pre[[f'lag_{i}' for i in range(1, lags + 1)]]
y_pre = df_daily_pre['total_usage']

# Train the Ridge regression model on the pre-COVID data
model_ridge = Ridge(fit_intercept=True)
model_ridge.fit(X_pre, y_pre)

# We will use the model trained on pre-COVID data to predict the bike usage during the COVID period
# For this, we need the last 'lags' days of the pre-COVID data to start the predictions
last_known_values = df_daily_pre['total_usage'][-lags:].values

# Generate predictions for the COVID period
predicted_bikes = []
for _ in df_daily_cov.index:
    current_features = np.array(last_known_values).reshape(1, -1)
    predicted_value = model_ridge.predict(current_features)[0]
    predicted_bikes.append(predicted_value)
    # Update the rolling window with the actual bike usage during COVID to simulate the prediction
    # in a more realistic way where the model is not aware of the pandemic
    last_known_values = np.append(last_known_values[1:], df_daily_cov['total_usage'].iloc[_])

# Plotting the actual vs predicted bike usage
plt.figure(figsize=(15, 5))
plt.plot(df_daily_cov['interval'], df_daily_cov['total_usage'], label='Actual COVID Period Rentals', color='blue')
plt.plot(df_daily_cov['interval'], predicted_bikes, label='Predicted Rentals If No COVID', color='red', linestyle='--')
plt.title('COVID Period: Actual vs Predicted Bike Rentals')
plt.xlabel('Date')
plt.ylabel('Total Bikes Rented')
plt.legend()
plt.show()

# Calculate RMSE and MAE between the actual and predicted values for the COVID period
rmse = mean_squared_error(df_daily_cov['total_usage'], predicted_bikes, squared=False)
mae = mean_absolute_error(df_daily_cov['total_usage'], predicted_bikes)
print(f'COVID Period RMSE: {rmse}')
print(f'COVID Period MAE: {mae}')

# Calculate the R^2 score between actual and predicted values
r_squared = r2_score(df_daily_cov['total_usage'], predicted_bikes)
print(f'R-squared (R^2) Score: {r_squared:.2f}')
```

```python
from sklearn.linear_model import Ridge,RidgeCV
from sklearn.model_selection import TimeSeriesSplit
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



df_daily_pre = pd.read_csv('daily_usage_pre_everyday.csv')
df_daily_cov = pd.read_csv('daily_usage_cov_everyday.csv')

# Convert the 'interval' columns to datetime and sort the data
df_daily_pre['interval'] = pd.to_datetime(df_daily_pre['interval'])
df_daily_cov['interval'] = pd.to_datetime(df_daily_cov['interval'])
df_daily_pre.sort_values(by='interval', inplace=True)
df_daily_cov.sort_values(by='interval', inplace=True)

# Create lagged features for the pre-COVID data
lags = 7  # Number of days for lagged features
for i in range(1, lags + 1):
    df_daily_pre[f'lag_{i}'] = df_daily_pre['total_usage'].shift(i)
df_daily_pre.dropna(inplace=True)

# Prepare the features and target variable for the pre-COVID data
X_pre = df_daily_pre[[f'lag_{i}' for i in range(1, lags + 1)]].values
y_pre = df_daily_pre['total_usage'].values

# Define the alphas for RidgeCV to test
alphas = np.logspace(-4, 4, 100)

# Create TimeSeriesSplit cross-validator
tscv = TimeSeriesSplit(n_splits=5)

# Instantiate the RidgeCV regressor
model_ridge_cv = RidgeCV(alphas=alphas, cv=tscv, scoring='neg_mean_squared_error')

# Fit the model
model_ridge_cv.fit(X_pre, y_pre)

# Get the best alpha value and corresponding MSE
best_alpha = model_ridge_cv.alpha_
best_mse = -model_ridge_cv.best_score_

print(f"The best alpha value found is: {best_alpha}")
print(f"The cross-validated MSE score for the best alpha: {best_mse}")

# Manually compute RMSE for each cross-validation fold
cv_rmse_scores = []
for train_index, test_index in tscv.split(X_pre):
    X_train, X_test = X_pre[train_index], X_pre[test_index]
    y_train, y_test = y_pre[train_index], y_pre[test_index]
    model = Ridge(alpha=best_alpha)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    cv_rmse_scores.append(np.sqrt(mse))

# Plot the RMSE scores
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cv_rmse_scores) + 1), cv_rmse_scores, marker='o')
plt.xlabel('Fold Number')
plt.ylabel('Root Mean Squared Error')
plt.title('RMSE for Each Fold in Time Series Cross-Validation')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


df_daily_pre = pd.read_csv('daily_usage_pre_everyday.csv')
df_daily_cov = pd.read_csv('daily_usage_cov_everyday.csv')

'''# Combine the datasets to scale them together
df_combined = pd.concat([df_daily_pre['total_usage'], df_daily_cov['total_usage']])

# Scale the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df_combined.values.reshape(-1, 1))'''

df_daily_pre.sort_values(by='interval', inplace=True)
df_daily_cov.sort_values(by='interval', inplace=True)

# Combine the 'total_usage' columns for scaling
df_combined = pd.concat([df_daily_pre['total_usage'], df_daily_cov['total_usage']], axis=0)

# Scale the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df_combined.values.reshape(-1, 1))

# Split the scaled data back into the original pre-COVID and COVID parts
scaled_pre = scaled_data[:len(df_daily_pre)]
scaled_cov = scaled_data[len(df_daily_pre):]

# Function to create a dataset to feed into the LSTM
def create_dataset(data, look_back=1):
    X, Y = [], []
    for i in range(len(data)-look_back-1):
        a = data[i:(i+look_back), 0]
        X.append(a)
        Y.append(data[i + look_back, 0])
    return np.array(X), np.array(Y)

look_back = 7  # Number of previous days the model will use to predict the next value
X, Y = create_dataset(scaled_data, look_back)

# Split the data into train and test sets
train_size = int(len(df_daily_pre) - look_back)
test_size = len(df_daily_cov) - (look_back + 1)
trainX, testX = X[:train_size], X[train_size:train_size+test_size]
trainY, testY = Y[:train_size], Y[train_size:train_size+test_size]

# Reshape input to be [samples, time steps, features]
trainX = np.reshape(trainX, (trainX.shape[0], trainX.shape[1], 1))
testX = np.reshape(testX, (testX.shape[0], testX.shape[1], 1))

# Define the LSTM model
model = Sequential()
model.add(LSTM(50, input_shape=(look_back, 1)))
model.add(Dropout(0.2))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model.fit(trainX, trainY, epochs=20, batch_size=1, verbose=1)

# Make predictions
train_predict = model.predict(trainX)
test_predict = model.predict(testX)

# Invert predictions to original scale
train_predict = scaler.inverse_transform(train_predict)
trainY = scaler.inverse_transform([trainY])
test_predict = scaler.inverse_transform(test_predict)
testY = scaler.inverse_transform([testY])

# Calculate metrics
test_rmse = np.sqrt(mean_squared_error(testY[0], test_predict[:,0]))
test_mae = mean_absolute_error(testY[0], test_predict[:,0])
test_r2 = r2_score(testY[0], test_predict[:,0])

# Plotting the actual vs predicted bike usage
plt.figure(figsize=(15, 5))

# Plot actual rentals for the COVID period
# Ensure that the 'interval' array matches the length of the 'testY' array
actual_dates = df_daily_cov['interval'].iloc[look_back + 1:look_back + 1 + len(test_predict)]
actual_usage = df_daily_cov['total_usage'].iloc[look_back + 1:look_back + 1 + len(test_predict)]

plt.plot(actual_dates, actual_usage, label='Actual COVID Period Rentals', color='blue')

# Plot predicted rentals if no COVID
# Make sure the dates array matches the length of the predictions
predicted_dates = df_daily_cov['interval'].iloc[look_back + 1:look_back + 1 + len(test_predict)]

plt.plot(predicted_dates, test_predict, label='Predicted Rentals If No COVID', color='red', linestyle='--')

plt.title('COVID Period: Actual vs Predicted Bike Rentals with LSTM')
plt.xlabel('Date')
plt.ylabel('Total Bikes Rented')
plt.legend()
plt.show()

# Metrics
print(f'COVID Period RMSE: {test_rmse}')
print(f'COVID Period MAE: {test_mae}')
print(f'R-squared (R^2) Score: {test_r2}')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Load your datasets
df_daily_pre = pd.read_csv('daily_usage_pre_everyday.csv')
df_daily_cov = pd.read_csv('daily_usage_cov_everyday.csv')

'''# Combine the datasets to scale them together
df_combined = pd.concat([df_daily_pre['total_usage'], df_daily_cov['total_usage']])

# Scale the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df_combined.values.reshape(-1, 1))'''

df_daily_pre.sort_values(by='interval', inplace=True)
df_daily_cov.sort_values(by='interval', inplace=True)

# Combine the 'total_usage' columns for scaling
df_combined = pd.concat([df_daily_pre['total_usage'], df_daily_cov['total_usage']], axis=0)

# Scale the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df_combined.values.reshape(-1, 1))

# Split the scaled data back into the original pre-COVID and COVID parts
scaled_pre = scaled_data[:len(df_daily_pre)]
scaled_cov = scaled_data[len(df_daily_pre):]

# Function to create a dataset to feed into the LSTM
def create_dataset(data, look_back=1):
    X, Y = [], []
    for i in range(len(data)-look_back-1):
        a = data[i:(i+look_back), 0]
        X.append(a)
        Y.append(data[i + look_back, 0])
    return np.array(X), np.array(Y)

look_back = 7  # Number of previous days the model will use to predict the next value
X, Y = create_dataset(scaled_data, look_back)

# Split the data into train and test sets
train_size = int(len(df_daily_pre) - look_back)
test_size = len(df_daily_cov) - (look_back + 1)
trainX, testX = X[:train_size], X[train_size:train_size+test_size]
trainY, testY = Y[:train_size], Y[train_size:train_size+test_size]

# Reshape input to be [samples, time steps, features]
trainX = np.reshape(trainX, (trainX.shape[0], trainX.shape[1], 1))
testX = np.reshape(testX, (testX.shape[0], testX.shape[1], 1))

# Define the LSTM model
model = Sequential()
model.add(LSTM(50, input_shape=(look_back, 1)))
model.add(Dropout(0.2))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model.fit(trainX, trainY, epochs=20, batch_size=1, verbose=1)

# Make predictions
train_predict = model.predict(trainX)
test_predict = model.predict(testX)

# Invert predictions to original scale
train_predict = scaler.inverse_transform(train_predict)
trainY = scaler.inverse_transform([trainY])
test_predict = scaler.inverse_transform(test_predict)
testY = scaler.inverse_transform([testY])

# Calculate metrics
test_rmse = np.sqrt(mean_squared_error(testY[0], test_predict[:,0]))
test_mae = mean_absolute_error(testY[0], test_predict[:,0])
test_r2 = r2_score(testY[0], test_predict[:,0])

# Plotting the actual vs predicted bike usage
plt.figure(figsize=(15, 5))

# Plot actual rentals for the COVID period
# Ensure that the 'interval' array matches the length of the 'testY' array
actual_dates = df_daily_cov['interval'].iloc[look_back + 1:look_back + 1 + len(test_predict)]
actual_usage = df_daily_cov['total_usage'].iloc[look_back + 1:look_back + 1 + len(test_predict)]

plt.plot(actual_dates, actual_usage, label='Actual COVID Period Rentals', color='blue')

# Plot predicted rentals if no COVID
# Make sure the dates array matches the length of the predictions
predicted_dates = df_daily_cov['interval'].iloc[look_back + 1:look_back + 1 + len(test_predict)]

plt.plot(predicted_dates, test_predict, label='Predicted Rentals If No COVID', color='red', linestyle='--')

plt.title('COVID Period: Actual vs Predicted Bike Rentals with LSTM')
plt.xlabel('Date')
plt.ylabel('Total Bikes Rented')
plt.legend()
plt.show()

# Metrics
print(f'COVID Period RMSE: {test_rmse}')
print(f'COVID Period MAE: {test_mae}')
print(f'R-squared (R^2) Score: {test_r2}')
```
